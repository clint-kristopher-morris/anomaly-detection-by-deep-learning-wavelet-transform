```python
"""
Add color coded weather bar for weather data
"""

import os
import cv2
import numpy as np
from project_tools import load_obj
df2dict_5min = load_obj('df2dict_5min')


path2dir = 'data/IMGS/scaleogram_im'
files = os.listdir(path2dir)
for data_file in files:
    
    im = cv2.imread(f'{path2dir}/{data_file}')
    
    color_data = df2dict_5min[data_file.split('---')[1].replace('.png','')]
    
    crop_img = im[24:240, 48:562]
    resized_image = cv2.resize(crop_img, (288, 288))
    output = np.zeros((315,288,3), dtype='uint8')
    for row in range(315):
        for col in range(288):
            if row < 288:
                output[row, col] = output[row, col] + resized_image[row, col]
            else:
                if color_data[col] == 1.0:
                    output[row, col] = [0,255,255]
                elif color_data[col] == 2.0:
                    output[row, col] = [0,0,255]
                else:
                    output[row, col] = [255,255,255]
    outt = cv2.resize(output, (500, 315))
    cv2.imwrite(f'data/IMGS/with_weather/{data_file}',outt)
```
